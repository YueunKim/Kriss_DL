```python
# 필요 모듈 임포트
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
import numpy
# 모델 정의
def create_model(optimizer='rmsprop', init='glorot_uniform'):
	model = Sequential()
	model.add(Dense(16, input_dim=8, kernel_initializer=init, activation='relu'))
	model.add(Dense(8, kernel_initializer=init, activation='relu'))
	model.add(Dense(1, kernel_initializer=init, activation='sigmoid'))
	model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
	return model
# 랜덤시드
seed = 123
numpy.random.seed(seed)
# 데이터 로드
dataset = numpy.loadtxt("/content/Pima_diabets_dataset.csv", delimiter=",", skiprows=1)
X = dataset[:,0:8]
Y = dataset[:,8]
# 모델 생성
model = KerasClassifier(build_fn=create_model)
```

```python
# Grid Search Hyperparameter 설정
optimizers = ['sgd','adam']
inits = ['normal', 'he_normal']
epochs = [50]
batches = [10, 20]
param_grid = dict(optimizer=optimizers, epochs=epochs, batch_size=batches, init=inits)
grid = GridSearchCV(estimator=model, param_grid=param_grid, verbose=1) 
# 파라미터별 학습 
grid_result = grid.fit(X, Y)

# 최적 파라미터 결과 출력
print("Best Score: %f with %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
params = grid_result.cv_results_['params']
for mean, param in zip(means, params):
	print("%f with: %r" % (mean, param))
```
