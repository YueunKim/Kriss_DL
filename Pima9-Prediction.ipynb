```python
# 필요 모듈 임포트
from keras.models import Sequential
from keras.layers import Dense
from keras.models import model_from_json
from keras.callbacks import ModelCheckpoint
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy
import os

# 결과 재현을 위한 난수 시드 정의
seed = 1
numpy.random.seed(seed)

# 데이터 로드
dataset = numpy.loadtxt("/content/drive/My Drive/USTxAI교육/실습/PimaIndian/Pima_diabets_dataset.csv", delimiter=",", skiprows=1)
X = dataset[:,0:8]
Y = dataset[:,8]
# 학습 집합과 테스트 집합 나누기
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.1, random_state=seed)

# 모델 생성
model = Sequential()
model.add(Dense(16, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# 모델 저장
model_save = model.to_json()
with open("simple_model.json", "w") as json_file:
    json_file.write(model_save)
```

```python
# 체크포인트 설정 및 웨이트 저장 설정
filepath="best_acc.h5"
checkpoint = ModelCheckpoint(filepath, monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')
callbacks_list = [checkpoint]

# 학습
model.fit(X_train, y_train, validation_split=0.33, epochs=50, batch_size=10, callbacks=callbacks_list, verbose=0)
```

```python
# 모델 및 웨이트 로딩
json_file = open("simple_model.json","r")
model_load = json_file.read()
json_file.close()
new_model = model_from_json(model_load)
new_model.load_weights("best_acc.h5")

# 로딩된 모델 및 웨이트로 결과 평가
new_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
scores = new_model.evaluate(X_test,y_test)
print("\n%s: %.2f%%" % (new_model.metrics_names[1], scores[1]*100))
```
